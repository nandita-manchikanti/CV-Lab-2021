# Q1:
1.
a. Take an image of the wall of your house and apply a sobel filter and count the number of edge pixels in the edge map

b. Do the same operation for the image of a tree ( capture your own image of the tree ) and report your observation on the two images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#from google.colab.patches import cv2_imshow
```

# Ques 1.a:
    


```python
def sobelfilter(s):
  img=cv2.imread(s)

  imggray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

  ret,imgBinary = cv2.threshold(imggray, 127, 255, cv2.THRESH_BINARY) 

  sobelx = cv2.Sobel(imgBinary, cv2.CV_64F, 0, 1, ksize=3)
  sobely = cv2.Sobel(imgBinary, cv2.CV_64F, 1, 0, ksize=3)
  sobelxy=cv2.bitwise_or(sobelx,sobely)
  sobelinverted=255-sobelxy


  fig, ax = plt.subplots(1,3,figsize=(20,20))
  ax[0].imshow(imggray,cmap='gray'),      ax[0].set_title('Gray')
  ax[1].imshow(imgBinary,cmap='gray'),    ax[1].set_title('Binary')
  ax[2].imshow(sobelinverted,cmap='gray'),ax[2].set_title('Sobel')

  edge_pixels=np.count_nonzero(sobelxy)
  print("Number of edge pixels:",edge_pixels)
    
```

```python
sobelfilter("tree.jpg")
sobelfilter("wall.jpg")
```
