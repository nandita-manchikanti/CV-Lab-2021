# Q2:
Write the python program which takes two polynomial as input and output the product of those two polynomials using convolution in spatial domain
(in O(n^2))


```python
import numpy as np
```

Funtion for convolution 

```python
def convolution(M1,M2):
  M2 = np.flip(M2)  #reverse of polynomial 
  # same as correlation
  li, = M1.shape  #length of first polynomial
  ls, = M2.shape  #length of second polynomial 
  lo = li + ls -1     #output matrix size      
  g = np.zeros([li+ls*2-2])
  g[ls-1:li+ls-1] = M1
  out = np.zeros([lo])
  for x in range(lo):
      C = np.multiply(g[x:x+ls],M2)
      out[x] = np.sum(C)
  return out , lo
```

Funtion to represent the array in the form of polynomial

```python
def polynomial(arr ,len):
  for i in reversed(range(len)):
    if (i==0):
      print(arr[len-1],end=" ");
    else:
      print(arr[len-i-1],"x^",i,"+",end=" ");  
```

The main code block to take input and produce the output 

```python
print("Enter the length of polynomial 1:") 
l1 = int(input())
print("Enter the length of polynomial 2:") 
l2 = int(input())
print("Enter values for polynomial 1:")
X1 = np.array([[]])
for i in range(l1):
  X1 = np.append(X1,int(input()))
print("Enter values for polynomial 2:")
X2 = np.array([[]])
for i in range(l2):
  X2=np.append(X2,int(input()))
finalpoly , l3 = convolution(X1,X2) 
print("Polynomial 1:")
print(polynomial(X1,l1))
print("Polynomial 2:")
print(polynomial(X2,l2))
print("Product of Polynomials :")
print(polynomial(finalpoly,l3))
```
